```python
dataframeLondonMerged = pd.merge(dataframeLondonUTD19, dataframeDetectors, on='detid', how='inner')
dataframeLondonMerged.head()
```

```python
unique_detectors = dataframeLondonMerged['detid'].unique()
random_detector = pd.Series(unique_detectors).sample(1).iloc[0]
print(f"Selected detector: {random_detector}")
dataframeDetector = dataframeLondonMerged[dataframeLondonMerged['detid'] == random_detector]
dataframeDetector = dataframeDetector.drop(columns=['detid'])
dataframeDetector.head()
```
