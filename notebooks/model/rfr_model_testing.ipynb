## RFR model functions testing

```python
import pandas as pd
import rfr_model as rm
```

### Load data

```python
# Read the data
train_data_path = '' # Insert train data path
test_data_path = '' # Insert test data path
train_data = pd.read_csv(train_data_path)
test_data = pd.read_csv(test_data_path)
train_data = train_data.drop(columns=['day', 'long', 'lat', 'pos', 'length'], axis=1)
test_data = test_data.drop(columns=['day', 'long', 'lat', 'pos', 'length'], axis=1)
```

### Use a subset of the train data

```python
train_head = train_data.head(10000).copy()
train_head
```

### Train RFR models

```python
save_path = '' # Insert save path for the RFR models
```

```python
rm.train_rfr_models(train_head, save_path)
```

```python
head_sensors = train_head.detid.unique().tolist()
```

```python
test_data_subset = test_data.copy()
test_data_subset = test_data_subset[test_data_subset['detid'].isin(head_sensors)]
test_data_subset
```

### Evaluate performance metrics

```python
evaluation = rm.evaluate_rfr_models(test_data_subset, save_path)
```

```python
print(f'Mean Aboslute Error: {evaluation.MAE.mean()}')
print(f'Mean Squared Error: {evaluation.MSE.mean()}')
print(f'Mean R2 Score: {evaluation.R2.mean()}')
```
